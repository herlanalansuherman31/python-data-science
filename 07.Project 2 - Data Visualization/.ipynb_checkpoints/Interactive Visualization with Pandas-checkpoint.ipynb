```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# reading the csv file
df_population_raw = pd.read_csv('population_total.csv')
# dropping null values
df_population_raw.dropna(inplace=True)
# making a pivot table
df_pivot = df_population_raw.pivot(index='year', columns='country',
                                        values='population')
# selecting some countries
df_pivot = df_pivot[['United States', 'India', 'China', 
                          'Indonesia', 'Brazil']]
# showing pivot table
df_pivot.head()
```

# Lineplot

```python
fig = px.line(
    df_pivot,
    x=df_pivot.index,
    y=df_pivot.columns,
    labels={'value':'Population','index':'Year'},
    title='Population (1955-2020)',
    template='seaborn'
)
fig.show()
```

# Barplot

```python
# selecting only one year (2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
# transpose dataframe from rows to columns or viceversa
df_pivot_2020 = df_pivot_2020.T
```

```python
fig = px.bar(
    df_pivot_2020,
    x=df_pivot_2020.index,
    y='2020',
    labels={'x':'Country','2020':'Population'},
    title='Population in 2020',
    template='ggplot2'
)
fig.show()
```

## Barplot grouped by “n” variables

```python
# selecting a few years
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
```

```python
fig = px.bar(
    df_pivot_sample,
    x=df_pivot_sample.index,
    y=df_pivot_sample.columns,
    labels={'value':'Population','index':'Year','variable':'Country'},
    title='Population Sample (per Year)',
    template='ggplot2'
)
fig.show()
```

# Piechart

```python
# changing column name 
df_pivot_2020.rename(columns={2020:'2020'}, inplace=True)
```

```python
fig = px.pie(
    df_pivot_2020,
    names=df_pivot_2020.index,   # nama negara
    values='2020',               # populasi tahun 2020
    title='Population in 2020',
    template='ggplot2'
)
fig.show()
```

# Boxplot

```python
fig = px.box(
    df_pivot,
    y='United States',
    labels={'United States':'Population'},
    title='Boxplot of US Population (1955-2020)',
    template='ggplot2',
)
fig.show()
```

```python
fig = px.box(df_pivot, y = df_pivot.columns, labels = {'value':'population'})
fig.show()
```

# Histogram

```python
fig = px.histogram(
    df_pivot,
    x=['United States','Indonesia'],
    title='Distribusi Populasi US vs Indonesia'

)
fig.show()
```

# ScatterPlot

```python
fig = px.scatter(
    df_pivot,
    x='Indonesia',
    y='United States',
    labels={'United States':'US Population','Indonesia':'Indonesia Population'},
    title='US vs Indonesia Population Scatter',
    template='ggplot2'
)
fig.show()
```
